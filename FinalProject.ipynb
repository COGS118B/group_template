# Proposed Solution

Our project aims to synthesize historical NFL game data and Super Bowl winners to identify key statistical predictors of future Super Bowl victories. We employed a multifaceted machine learning approach to achieve this objective:
Feature Selection via Random Forest Classification: Initially, we utilized a random forest classifier to discern the most significant features within our dataset, effectively isolating the critical variables that contribute to a team's success.
Exploratory Clustering with K-Means: Leveraging the identified key features, we applied k-means clustering to uncover patterns and similarities among teams, focusing on these influential factors. To determine the optimal number of clusters (k), we employed both the elbow method and silhouette scores, which suggested a division into four distinct clusters.
Data Preparation and SVM Classification: We partitioned our dataset into training (80%) and testing (20%) subsets, preparing our data for model training. Subsequently, we implemented a Support Vector Machine (SVM) classifier to predict the cluster assignments of new data points, leveraging the patterns identified during the clustering phase.
Model Evaluation: We assessed the accuracy of our SVM model to ensure its effectiveness in correctly predicting cluster assignments, thereby validating our methodology.
Hierarchical Clustering: In addition to k-means clustering, we explored hierarchical clustering to observe the data in a different light. This method allowed us to visualize each data point as an individual cluster initially, with a dendrogram illustrating the sequence of cluster mergers and the distances at which they occurred.
Dimensionality Reduction with PCA: We performed Principal Component Analysis (PCA) to condense the feature space into two dimensions, facilitating a visual exploration of the dataset's global relationships through a scatter plot.
Linear Regression Analysis: Lastly, we conducted linear regression to predict winning outcomes based on the diverse features identified. This analysis aimed to quantify the impact of each feature on a team's likelihood of winning.
Through this comprehensive analysis, we strive to gain deeper insights into the factors that significantly influence a team's prospects of becoming Super Bowl champions, enhancing our understanding of football analytics.



# Evaluation Metric

For our evaluation metrics, we knew we could use a few measurements to evaluate both the benchmark model and the solution model. For our classification model, where we are making our predictions for teams that will win the superbowl, we can use accuracy as the most straightforward evaluation metric, and calculate the proportion of correct predictions (both true positives and true negatives) out of all predictions. We can also use precision and recall to  ​​measure the proportion of positive identifications that were actually correct, and the proportion of actual positives that were identified correctly. These measurements will be particularly useful because incorrectly predicting a team as a Super Bowl winner when it's not is less desirable than missing a potential winner. For our unsupervised analysis, we’re thinking about using clustering for team profiling. Using K-means or hierarchical clustering to group teams into clusters based on their performance metrics. These clusters can help identify patterns of team performance that correlate with Super Bowl success. For example, teams within a cluster characterized by a strong defense and decent offensive scores might have a higher likelihood of winning the Super Bowl. For this model, we can use the silhouette score as our evaluation metric. This metric measures how similar an object is to its own cluster compared to other clusters. This score ranges from -1 to 1, where a high value indicates that the object is well matched to its own cluster and poorly matched to neighboring clusters. In the context of team profiling, a high silhouette score would indicate good clustering and that teams within a cluster are similar to one another and distinct from teams in other clusters.

# Elbow Method and Silhouette Score

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import silhouette_score


# Prepare features for K-Means clustering
features = ['score_away', 'score_home', 'first_downs_away', 'first_downs_home', 'yards_home', 'yards_away', 'fumbles_away', 'fumbles_home', 'interceptions_away', 'interceptions_home','winner', 'third_down_comp_away', 'third_down_comp_home']  # Updated to include more features
X = data[features] 

# Normalize features for clustering
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

wcss = []
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
    

plt.figure(figsize=(10, 5))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method For Optimal Cluster Number')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()


silhouette_scores = [] 
for n_cluster in range(2, 11):
    kmeans = KMeans(n_clusters=n_cluster).fit(X_scaled)
    silhouette = silhouette_score(X_scaled, kmeans.labels_)
    silhouette_scores.append(silhouette)
    print("For n_clusters = {}, silhouette score is {})".format(n_cluster, silhouette))
    
plt.figure(figsize=(10, 5))
plt.plot(range(2, 11), silhouette_scores, marker='o', linestyle='--')
plt.title('Silhouette Score For Optimal Cluster Number')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')
plt.show()
```

From the Elbow Method, we can see that our ideal cluster number is 4. Also, our Silhouette Score indicates that 4 is a good cluster number.

## Limitations

Our limitations on this project mostly stem from the nature of our dataset and the complexity of the prediction we are trying to make. More comprehensive data could potentially change the nature of the problem. Including additional features such as player injuries, team morale, coaching strategies, weather conditions, and mid-season trades, etc. could provide a more holistic view. However, collecting and integrating this data would be pretty challenging. Also, the granularity of the data can significantly impact model performance. For instance, player-level statistics might offer more insights than team-level aggregates, but they also introduce more complexity. 
Another huge limitation is that the outcome of sports events, such as the Super Bowl, is influenced by many factors, some quantifiable and others intangible. This complexity makes it inherently difficult to predict outcomes with high accuracy. Not to mention the risk of overfitting, especially when the dataset is not large enough compared to the number of features or when the model is too complex. This is something we experienced a little in our models and something that would absolutely happen if we introduced more complexity in our feature selection. This will most likely lead to models that perform well on historical data but fail to generalize to future games. Given the complexity and unpredictability of sports outcomes, exploring a wider range of hyperparameters could potentially improve model performance. However, this requires significant computational resources and time. If resources and time permitted we would further explore automated hyperparameter optimization techniques like grid search, random search, or Bayesian optimization.


## Ethics & Privacy

There aren’t any super obvious ethical concerns with our project, but as with every ML project the concern of model bias and fairness. If we want it to be considered fair then the model should avoid reinforcing biases present in the data, which could unfairly advantage or disadvantage certain teams based on factors unrelated to their performance. This is not a big concern for us because all of our data is performance related. However, there could be external factors influencing certain variables that are unknown to us. What is within our power is to continuously monitor the data and employ any bias mitigation techniques we have available.
	Also, given the popularity of betting around sporting events like the Super Bowl, models predicting outcomes could be misused for gambling purposes, raising ethical concerns around fairness and the potential for addiction. Even though our model isn’t super accurate in its predictions, if the model was accurate, someone could argue that it gives sports betters an unfair advantage over betters that don’t use the model.
